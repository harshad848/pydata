```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import OneHotEncoder,OrdinalEncoder,StandardScaler,MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LogisticRegression
```

```python
df=pd.read_csv('dataset/titanic.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df.drop(['PassengerId','Ticket','Fare','Cabin','Name'],inplace=True,axis=1)
```

```python
df
```

```python
df.fillna(method='ffill',inplace=True)
```

```python
df
```

```python
x=df.iloc[:,1:]
```

```python
y=df.iloc[:,0]
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
tr=ColumnTransformer(transformers=[
    ('t1',OneHotEncoder(),['Sex','Embarked'])
],remainder='passthrough'
                    
                    )
```

```python
a=tr.fit_transform(x_train
                  
                  )
```

```python
a
```

```python
b=tr.fit_transform(x_test)
```

```python
b
```

```python
log=LogisticRegression()
```

```python
log.fit(a,y_train)
```

```python
y_pred=log.predict(b)
y_pred
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
print('Accuracy matrix =',accuracy_score(y_test,y_pred))
```

```python
pd.DataFrame(confusion_matrix(y_test,y_pred))
```
