```python
q1=df['Age'].quantile(0.25)
q1
```

```python
q2=df['Age'].quantile(0.75)
q2
```

```python
iqr=q2-q1
iqr
```

```python
ll=q1-1.5*iqr
ll
```

```python
ul=q2+1.5*iqr
ul
```

```python
df[df['Age']>ul]
```

```python
df['Age']=np.where(df['Age']>ul,
                  ul,
                  np.where(df['Age']<ll,
                          ll,
                          df['Age']))
df
```
