```python
from pandas import DataFrame
df=DataFrame(
            [
                ['Aryan','Singh','Full Time','administraction',2,20000],
                ['Rohan','Agarwal','intern','Technical',3,50000],
                ['Rjya','Shah','Full Time','administraction',5,100000],
                ['Yash','Bhaita','Part Time','Technical',7,100000],
                ['Siddhant','Khanna','Full Time','Management',6,200000],
            ]
            ,columns=['Frist Name','Last Name','Type','Department','YoE','Salary']
            )
df
```

```python
import pandas as pd
table=pd.pivot_table(df,index='Type')
table
```

```python
import pandas as pd
table=pd.pivot_table(df,index='Type',columns='Department')
table
```

```python
import pandas as pd
table=pd.pivot_table(df,index='Type',columns='Department',values='Salary',aggfunc='sum')
table
```
