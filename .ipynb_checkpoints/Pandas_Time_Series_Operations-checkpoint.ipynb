*The Series include the data and time for every record and can fetch the record of dataframe.We can find out the data within certain range of data and time by using pandas module named time series

```python
import numpy as np
import pandas as pd
from datetime import datetime
time=pd.date_range(start='10/1/2022',end='11/1/2022',freq='Min')
print(time)
```

```python
time=pd.date_range(start='10/1/2022',end='11/1/2022',freq='Min')
df=pd.DataFrame(time,columns=['DateTime'])
print(df)
```

```python
time=pd.date_range(start='10/1/2022',end='11/1/2022',freq='Min')
df=pd.DataFrame(time,columns=['DateTime'])
df['DATA']=np.random.randint(0,100,size=(len(time)))
#df
df.head(20)
```
